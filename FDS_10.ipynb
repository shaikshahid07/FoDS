Experiment 1

```python
import numpy as np
import matplotlib.pyplot as plt

student_scores = np.array([
    [85, 78, 90, 88],
    [76, 82, 85, 80],
    [92, 88, 91, 95],
    [80, 75, 84, 82]
])

subjects = ["Math", "Science", "English", "History"]

avg = np.mean(student_scores, axis=0)

print("Average Scores")
for i in range(len(subjects)):
    print(subjects[i], ":", avg[i])

print("Highest Average Subject:", subjects[np.argmax(avg)])

plt.figure(figsize=(6,4))
plt.bar(subjects, avg)
plt.title("Average Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

EXPERIMENT 4

```python
import numpy as np
import matplotlib.pyplot as plt

sales_data=np.array([50000,60000,70000,90000])

total=np.sum(sales_data)

increase=((sales_data[3]-sales_data[0])/sales_data[0])*100

print("Total Sales:",total)
print("Percentage Increase:",increase,"%")

quarters=["Q1","Q2","Q3","Q4"]

plt.figure(figsize=(6,4))
plt.plot(quarters,sales_data,marker="o")
plt.title("Quarterly Sales")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

EXPERIMENT 6

```python
import numpy as np
import matplotlib.pyplot as plt

prices=np.array([50,80,120,40])
quantity=np.array([2,1,3,4])

discount=10
tax=5

subtotal=np.sum(prices*quantity)

discount_amount=subtotal*(discount/100)

after_discount=subtotal-discount_amount

tax_amount=after_discount*(tax/100)

total=after_discount+tax_amount

print("Subtotal:",subtotal)
print("Final Total:",total)

labels=["Subtotal","Discount","Tax","Final"]

values=[subtotal,discount_amount,tax_amount,total]

plt.figure(figsize=(6,4))
plt.bar(labels,values)
plt.title("Bill Summary")
plt.show()
```

EXPERIMENT 7

```python
import pandas as pd
import matplotlib.pyplot as plt

order_data=pd.DataFrame({
"CustomerID":[101,102,101,103,102,101],
"OrderDate":["2024-01-02","2024-01-05","2024-01-08","2024-01-10","2024-01-15","2024-01-20"],
"Product":["Laptop","Phone","Laptop","Tablet","Phone","Mouse"],
"Quantity":[2,1,3,2,4,5]
})

order_data["OrderDate"]=pd.to_datetime(order_data["OrderDate"])

print(order_data.groupby("CustomerID").size())

print(order_data.groupby("Product")["Quantity"].mean())

print("Earliest:",order_data["OrderDate"].min())

print("Latest:",order_data["OrderDate"].max())

order_data.groupby("CustomerID").size().plot(kind="bar")
plt.title("Orders per Customer")
plt.ylabel("Orders")
plt.show()
```

EXPERIMENT 9

```python
import pandas as pd
import matplotlib.pyplot as plt

property_data=pd.DataFrame({
"PropertyID":[1,2,3,4,5],
"Location":["A","B","A","C","B"],
"Bedrooms":[3,5,6,4,5],
"Area":[1200,2500,2200,1800,2600],
"Price":[300000,500000,450000,350000,600000]
})

print(property_data.groupby("Location")["Price"].mean())

print(property_data[property_data["Bedrooms"]>4].shape[0])

print(property_data.loc[property_data["Area"].idxmax()])

property_data.groupby("Location")["Price"].mean().plot(kind="bar")
plt.title("Average Price by Location")
plt.ylabel("Price")
plt.show()
```
